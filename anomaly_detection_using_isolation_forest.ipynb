**Fitting Isolation Forest**

Isolation Forest is an unsupervised learning algorithm that works detects anomalies by isolating them

We will check  whether our Model is able to detect the anomalous points.

```python
model=IsolationForest(contamination=0.004)
model.fit(df[['value']])
```

Build an ensemble of Isolation trees for the dataset.An anomaly has shorter average length on the isolation Trees.It isolates object by randomly selecting some features but here we have only one feature.It isolates by splitting the values between maximum and minimum range as it splits we see the anomalous point that stands apart.

Setting Hyperparameter called contamination to 0.004 i.e. the proportion of outliers that we want to identify. We set it using domain knowledge ,we should have an idea of what percentage of the data is anomalous beforehand to get a better prediction which is one of the limitations of isolation forest.It ranges between 0 to 0.5



```python
model.get_params() 
```

automatically sets default n_estimators=100 which is nothing but the number of trees.

max_features=1.0 i.e it takes all features as only one feature here.

Since contamination level is set predict will give The output from model as +1 or -1 when -1 that point is an anomaly.

```python
df['outliers']=pd.Series(model.predict(df[['value']])).apply(lambda x: 'yes' if (x==-1) else 'no' )
#lambdify x and if x==-1 then anomaly else no and added column outliers to same dataset
```

```python
df.query('outliers=="yes"') #did not detect christmas and thanksgiving
```

It did not detect christmas and thanksgiving
We have multiple kinds of outliers some outliers way outside the typical range known as **global outliers** but some are only **contextual outliers**(outliers for the business aspects but not for the data)

**Isolation forest without contamination hyperparameter**

```python
model=IsolationForest()  #automatic contamination will be detected by model 
model.fit(df[['value']])
```

```python
score=model.decision_function(df[['value']])   #Average anomaly score of X of the base classifiers.
#  The anomaly score of the input samples.The lower, the more abnormal. Negative scores represent outliers,positive scores represent inliers.
```

```python
score
```

```python
plt.hist(score,bins=50)  #histogram of scores
```

Lot of points on higher end ,but values on lower end of histogram are probable outliers.We can set a threshold less than which the point will be considered a outlier.  example -0.15 ,-0.20

```python
#adding score as a column to data
df['scores']=score
df
```

```python
df.query('scores<-0.20')  #all scores below -0.20 ,from histogram selected this 
```

Detected  new year but did not detect christmas and thanksgiving.We did not feed any intelligence to model to help detect outliers better which can be done using feature engineering.
We using non time series model to detect time series ouliers.

**As seen from the diagram contextual anomaly is in june as the normal pattern as compared with jan-march quarter is disrupted. A global anomaly or point anomaly would be easily visible on the much higher/lower end.**

    Thanksgiving-2014-11-27

```python
#Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length). The index can replace the
#existing index or expand on it.
df_final.set_index('timestamp').loc['2014-11-02']  #all data for this particular anomalous date NYC MARATHON
```

Value very higher than average at 1am in the night. This can be a global outlier or contextual outlier. Let's check on another  anomalous dates to compare.

```python
df_final.set_index('timestamp').loc['2015-01-26':'2015-01-27'] #SNOW BLIZZARD DATES DATA
```

We see very few taxi's running on that day compared to average value particularly due to snow blizzard

```python
df_final['day']=df.timestamp.dt.weekday   #converting the weekday column to numeric and adding new column ,monday=0,tuesday =1 ,wednesday=2 and so on..
df_final         #bcse need numeric data to feed to model categorical cannot be fed
```

```python
data=df_final[['value','hour','day']]   #slicing only 
data.head()
```

```python
model2=IsolationForest(contamination=0.005,max_features=3,max_samples=0.8,n_estimators=200)
model2.fit(data)   #max 80% samples used in each tree ,max_features in each tree =3,# of  trees 200
```

**Isolation Forest profiles anomalies instead of normal data. Contamination level controls threshold for decision function(how many anomalies want to consider top .5% of anomalies),we can optionally leave it be and set to auto.**

```python
df_final['outliers']=pd.Series(model2.predict(data)).apply(lambda x: 'yes' if (x==-1) else 'no')
#output of model.predict will be +1 for normal data or -1 for outliers, 
#so we creating column outliers to store outliers as yes and normal point as no
```

```python
df_final.query('outliers=="yes"')
```

This model was able to detect contextual outliers as well like christmas 25th december

```python
fig=px.scatter(df_final,x='timestamp',y='value',color='outliers',hover_data=['weekday','hour','value_avg'],  title='NYC Taxi Demand')
#weekday,hour,value_avg as a tooltip and outliers with color
fig.update_xaxes(rangeslider_visible=True)  #adding slider to stretch the line chart to see monthly to hourly pattern  
fig.show()  #shows outliers and when hover on them we can see details
```

We can see a lot of inlying contextual outliers  (compared to avg_value)

**Deciding Threshold**

```python
score=model2.decision_function(data)   #Average anomaly score of X of the base classifiers for data
```

```python
score
```

```python
plt.hist(score,bins=50) #plotting histogram of scores
```

Most values towards right,left values seem anomalous points .We can draw a cutoff (threshold). Here in this case we chose -0.02

```python
df_final['scores']=score
```

```python
df_final
```

```python
df_final.query('scores<-0.02')  #from histogram we can take threshold as -0.02
```

Here,we see that we got the anomalous hours for 26th and 27th January, christmas,new year and NYC Marathon were also detected.
Thanksgiving was still not detected however there are limitless number of combinations with feature engineering,contamination/ threshold which can be tried to spot the outliers.
